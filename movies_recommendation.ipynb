```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
selected_features = ['genres','keywords','tagline','cast','director','title']
print(selected_features)
```

```python
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
combined = movies_data['genres'] + ' ' + movies_data['keywords'] + ' ' + movies_data['tagline'] + ' ' + movies_data['cast'] + ' ' + movies_data['director'] + ' ' + movies_data['title']
```

```python
print(combined)
```

```python
vecorizer = TfidfVectorizer()
```

```python
vectors = vecorizer.fit_transform(combined)
```

```python
print(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
movie_name = input('Enter the movie name: ')
```

```python
all_titles = movies_data['title'].tolist()
print(all_titles)
```

```python
close_match = difflib.get_close_matches(movie_name, all_titles)
print(close_match)
```

```python
close_match = close_match[0]
print(close_match)
```

```python
index_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_movie)
```

```python
similarity_score = list(enumerate(similarity[index_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
print('Movies suggested for you: \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<20):
    print(i , '.', title_from_index)
    i+=1
```

```python
movie_name = input("Enter the movie name: ")

all_titles = movies_data['title'].tolist()

close_match = difflib.get_close_matches(movie_name, all_titles)

close_match = close_match[0]

index_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

print('Movies suggested for you: \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<20):
    print(i , '.', title_from_index)
    i+=1
```
